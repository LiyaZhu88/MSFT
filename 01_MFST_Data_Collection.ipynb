```python
#Step 1 Install yfinance and inport python libraries
!pip install yfinance
import yfinance as yf
import pandas as pd

#Step 2 Select Stock
ticker="MSFT"
stock = yf.Ticker(ticker)

#Step 3 Data Collection: Day data
rawdata_day = stock.history(interval="1d", start="2010-1-1", end="2023-1-1")
rawdata_day.to_csv('raw_data_day.csv')

#Step 4 Data Collection: 5 minutes data
rawdata_5min= stock.history(interval='5m')
rawdata_5min.to_csv('raw_data_5min.csv')
```
